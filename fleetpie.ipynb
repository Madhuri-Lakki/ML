```python
from reportlab.platypus import SimpleDocTemplate
from reportlab.graphics.charts.legends import Legend as legend
import reportlab.lib.colors as colors
from matplotlib import pyplot as plt
plt.rcParams['legend.handlelength'] = 1
plt.rcParams['legend.handleheight'] = 1.125
plt.rcParams['legend.numpoints'] = 1
#from reportlab.lib.colors import yellow, green, orange, HexColor
def main():
    data = [11,1,88]
    ##my_title = String(170, 40, 'My Pie Chart', fontSize=14)
    label1 = ['Running','Idling','Stopped',]
    cl = ['green', 'yellow', 'orangered']
    plt.pie(data, autopct = '%2.1f%%', colors = cl, startangle = 40, shadow = True,explode=(0.1,0,0))
    ax1 = plt.subplot(111)
    b1, =ax1.plot([], marker="s", markersize=10, linestyle="", label='Running',color='green')
    b2, =ax1.plot([], marker="s", markersize=10, linestyle="", label='Idling',color='yellow')
    b3, =ax1.plot([], marker="s", markersize=10, linestyle="", label='Stopped',color='orangered')
    ax1.legend(loc='upper left', bbox_to_anchor=(-0.2, 1.00), shadow=True, ncol=1,handles=[b1, b2, b3])
    ##plt.legend.defaultColors = [green, yellow, orangered]
    ##plt.legend()
    ##doc = SimpleDocTemplate('flowable_with_chart.pdf')
    fig = plt.savefig('/Users/lakkireddy.madhuri/reportlabs/piechrt.jpg',transparent=False)
    ##elements = []
    plt.show()
    ##elements.append(fig)
    ##doc.build(elements)
    
if __name__ == '__main__':
    main()
    
```
